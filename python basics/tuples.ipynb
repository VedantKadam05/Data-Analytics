```python
## Tuples are ordered collection of items which are immutable (cannot be changed after creation).
## creating a tuple 

tuple=(1,'vedant','yui')
print(tuple)
print(type(tuple))
```

```python
del tuple
```

```python
numbers = tuple([1,3,6,8,0]) ## a list converted into tuple
print(numbers)
```

```python
## Accessing tuple elements
print(numbers[0])
print(numbers[-1])
```

```python
numbers[0:4] ## slicing
```

```python
## tuples operation -- concatenation
tup1=(1,'kadam','talav','buddha')
tup2=('dhamma','goenka','2')
tup3=tup1+tup2
print(tup3)
```

```python
tup3 * 3 ## directly does print tup2 three times, if we want to really multiplyby 3 then we have to take a single tuple and use a iterator
```

```python
## Immutable nature of tuples

## tuples are immutable -- they cannnot be changed


tuple=(1,2,3,4,5,6,7,8,9)
tuple[0]=1
print(tuple)
```

```python
## Tuple Methods --} count,index
print(numbers.count(1)) ## print the number of times "1" is present

print(numbers.index(3)) ## returns the index of the value
```

```python
## packing and unpacking tuple
## packing
packed_tuple = 1,"hello",2
print(packed_tuple)
```

```python
## unpacking
a,b,c=packed_tuple ## see that a,b,c are assigned respectively to elements in packed_tuple
print(a)
print(b)
print(c)
```

```python
## unpacking with *
numbers=(1,2,3,4,5,7)
first,*second,last=numbers
print(first)
print(second) ## what this does is start with 2nd no. and goes till second last no. and print it
print(last)
```

```python
## Nested tuple
lst=[[1,2,3],[4,5,6],[7,"hello",'vedant']]

lst[0][2]=6 ## [0] is 1st list inside of lst and [2] is the third element(indexing starts with 0) 
print(lst[0])

print(lst)
```
