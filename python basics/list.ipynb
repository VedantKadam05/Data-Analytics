##### list comprehension
Basic syntax [expression for item iterable]

with conditional logic [expression for item in iterable if condition]

Nested List comprehension [expression for item1 in iterable1 for item2 in iterable2]

```python
## basic list comprehension
square=[num**2 for num in range(10)]
print(square)
```

```python
##list comprehension with condition
lst=[]
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)        
```

```python
##nested list comprehension
lst1=[1,2,3]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
## list comprehension with function cells

words = ["hello","world","python","programming"]
lengths = [len(word) for word in words]  #len(word) ----} len() is a built in function which gives the length of the string
print(lengths)
```

## Real life example of list
1.create to do list 

```python
to_do_list=["buy groceries","clean the house","pay bills"]

#adding task to list
to_do_list.append("go for a run")
to_do_list.append("read a book")

#removing a completed task
to_do_list.remove("clean the house")

#checking the task in list
if "pay bills" in to_do_list:
    print("dont forgot to pay bills")

for task in to_do_list:
    print(f"-{task}")   
```

# organizing student grade

```python
student_grade=[26,76,98,65,3,32,9,65,98,13,76]
#adding a new grade to list
student_grade.append(89)

#finding average grade 
average_grade = sum(student_grade)/len(student_grade)

print(f"average grade: {average_grade:.2f}")

#finding highest and lowest grades
highest_grades = max(student_grade)
lowest_grades = min(student_grade)

print(highest_grades)
print(lowest_grades)
```
