```python
##if statement
age=20
if age>=20:
    print("true")
```

```python
##if else
age=17
if age>=18:
    print("true")
else:
    print("false")    
```

```python
## elif
age=18
if age<13:
    print("you are a child")
elif age<18:
    print("you are a teenager")
else:
    print("you are adult")        
```

```python
## nested conditional statemen
days=int(input("enter number of days"))

if days==365:
    print("this is not a leap year")

elif days==366:
    print("this is a leap year")

else:
    print("invalid number of years")
```

```python
## simple calculator program
operator=(input("enter the operator add/sub/mul/div"))
num1=float(input("enter first number"))
num2=float(input("enter second number"))

if operator == '+':
    print(num1+num2)
if operator == '-':
    print(num1-num2)
if operator == '*':
    print(num1*num2)
if operator == '/':
    print(num1/num2)
else:
    print("invalid number")


```
