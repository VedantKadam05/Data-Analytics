```python
## dictionaries are unordered collections of items. they store data in 'key value' pairs. keys must be unique and immutable (e.g. strin,numbers,tuples)
```

```python
## creating dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
student={"name":"vedant","age":20,"grade":98}
print(student)
```

```python
## accessing dictionary elements
print(student['grade'])
print(student['name'])

## accessing using get() method
print(student.get('age'))
print(student.get('last_name',"not available"))
```

```python
## dictionaries are mutable, you can add,update,delete elements
print(student)
```

```python
student["age"]=33 ## updated the value
print(student)
student["address"]="india" ## added a new key and value
print(student)
```

```python
## deleting a key
del student["grade"]
print(student)
```

```python
## iterating over dictionary 
## iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## iterating over values
for value in student.values():
    print(keys)
```

```python
## iterating over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## dictionary comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## merge 2 dictionaries into one
dict1={"a":1,"b":2}
dict2={"c":3,"d":4}

merged_dict={**dict1,**dict2} ## ** is a any key-value pair gets appended into dictionaries
print(merged_dict)
```
