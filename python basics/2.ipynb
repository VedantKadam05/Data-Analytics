```python
## simple calculator
num1=float(input("what is the num1"))
num2=float(input("what is the num2"))

sum = num1 + num2
difference = num1 - num2
product = num1 * num2
division = num1 / num2
modulo = num1 % num2   # gives the remainder after the division

print("sum",sum)
print("difference",difference)
print("product",product)
print("division", division)
print("modulo",modulo)
```
