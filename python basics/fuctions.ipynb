```python
## variable length arguments
## positional and keywords arguments

def print_numbers(*vedant): ## *vedant is positional argument---} takes all the arguments given by user # *args should be written in place of *vedant -- good practice 
    for number in vedant:
        print(number)
```

```python
## keyword argument

def print_details(**kwargs):  ## are in the form of key value pairs
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="vedant",age="20",country="india")
```

```python
## examples 
#1.temp conversion
def convert_temp(temp,unit):
    '''this function converts the temp btween c to f'''
    if unit=='C':
        return temp* 9/5 + 35 #c to f
    elif unit =='F':
        return (temp-32)*5/9  #f to c
    
print(convert_temp(78,'F'))    
    
```

```python
#2. password strength checker
def is_strong_password(password):
    '''check if password is strong or not'''
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password): ## to see weather weather digit is present or not, hence we used char.digit() and loop to iterater through all character
        return False
    if not any(char.islower()for char in password):
        return False
    if not any(char.isupper()for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True

## calling the function
print(is_strong_password("Vedant@789"))
print(is_strong_password("Vedant@"))
```

```python
# check if string is palindrome
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

print(is_palindrome("A man a plan a canal panama"))
print(is_palindrome("this is india"))
```

```python
# calculate the factorial of number using recursion
def factorial(n):
    if n==0:
        return 1
    else:
        return n * factorial(n-1)
    
print(factorial(6))    
```
