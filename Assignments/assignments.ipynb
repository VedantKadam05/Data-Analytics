```python
#assignment 12 factorial calculation
num=int(input("enter a number of which you want factorial"))
result=1

for i in range(1,num+1): # num+1 is necessary to iterate till num in python ex (1,4)--} 1,2,3 only goes till 3 and not 4
    result=result*i
    print(result)
    
```
